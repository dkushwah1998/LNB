```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#static plot
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()
```

```python
plt.plot(x,np.sin(x),'--')
plt.plot(x,np.cos(x),'-')
plt.xlabel('100 Real Numbers Between 1 and 10')
plt.ylabel('Sin and Cos of x')
plt.title('Figure displaying the trignometric ratios')

plt.show()
```
