```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([0,2,4,6,8])
y = np.array([0,4,8,12,16])

p = np.array([0,2,4,6,8])
q = np.array([0,6,12,18,24])
```

```python

plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(p,q)
plt.suptitle('XYZ')
plt.show()
```
