```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]

slp_time = [6,8,6,5,10]
work = [7,6,7.5,12,6]

infot = [8,6,5,2,8]
plt.plot([],[],label = 'Sleeping',color = 'm')
plt.plot([],[],label = 'work',color = 'c')
plt.plot([],[],label = 'infot',color = 'r')
plt.stackplot(days,slp_time,work,infot,colors = ['m','c','r'])
plt.xlabel('days')
plt.ylabel('Activity')
plt.title('Distribution of daily activity')
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]

slp_time = [6,8,6,5,10]
work = [7,6,7.5,12,6]

infot = [8,6,5,2,8]

plt.stackplot(days,slp_time,work,infot,colors = ['m','c','r'],labels = ['sleeping','work','infot'])
plt.xlabel('days')
plt.ylabel('Activity')
plt.title('Distribution of daily activity')
plt.legend()
plt.show()
```

```python
slices = [6,7,8]
col = ['m','c','r']
plt.pie(slices,colors = col, labels = ['sleeping','work','entertainment'],autopct='%1.1f%%',explode=(0,0.2,0.05))
plt.title('Activity breakdown for monday')
plt.show()
```
